```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Open the CSV file
df = pd.read_csv("/Users/Michelle Wong/Desktop/population.csv")

# Filter the data for African countries in the year 2010
df_2010_africa = df[(df['year'] == 2010) & (df['continent'] == 'Africa')]

# Calculate the total population for all African countries in 2010
total_population_africa_2010 = df_2010_africa['population'].sum()

# Display the total population for Africa in 2010
print("Total population of African countries in 2010: " + str(int(total_population_africa_2010)) + " million")

# Sort the data by population in descending order
df_2010_africa_sorted = df_2010_africa.sort_values('population', ascending=False)

# Normalize the population data to the range [0, 1] for colormap
norm = mcolors.Normalize(vmin=df_2010_africa_sorted['population'].min(), 
                         vmax=df_2010_africa_sorted['population'].max())

# Create the plot
plt.figure(figsize=(12, 6))
ax = plt.gca()  # Get current axes

# Use the colormap to apply a color gradient
cmap = plt.cm.viridis

# Create the bar plot with color based on population using the colormap
bars = ax.bar(df_2010_africa_sorted['country name'], 
              df_2010_africa_sorted['population'], 
              color=cmap(norm(df_2010_africa_sorted['population'])))

# Title and labels
ax.set_title('Population of African Countries (2010)', fontsize=14, color='white')
ax.set_ylabel('Population (Millions)', fontsize=12, color='white')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, color='white')
plt.yticks(color='white')

# Set background color of the entire figure
plt.gcf().set_facecolor('black')

# Set the background color of the axes (where the bars are)
ax.set_facecolor('black')

# Add horizontal gridlines with light grey color
ax.yaxis.grid(True, color='lightgrey', linestyle='--', linewidth=0.5)

# Disable the vertical gridlines
ax.xaxis.grid(False)

# Adjust layout to avoid overlap and extra spaces
plt.tight_layout()

# Show the plot
plt.show()
```
