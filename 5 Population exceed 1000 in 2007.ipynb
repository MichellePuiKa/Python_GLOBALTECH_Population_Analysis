```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt

# Open the CSV file
df = pd.read_csv("/Users/Michelle Wong/Desktop/population.csv")
    
# Filter the data for the year 2007
df_2007 = df[df['year'] == 2007]

# Identify the countries with populations exceeding 1000 million
df_above_1000 = df_2007[df_2007['population'] > 1000]

# Convert the population column to integer (removes decimal places)
df_above_1000['population'] = df_above_1000['population'].astype(int)

print("Countries with populations exceeding 1000 million in 2007:")
print(tabulate(df_above_1000[['country name', 'population']], headers=['Country', 'Population (Million)'], tablefmt='pretty', showindex=False))

import pandas as pd
import matplotlib.pyplot as plt

# Open the CSV file
df = pd.read_csv("/Users/Michelle Wong/Desktop/population.csv")
    
# Filter the data for the year 2007
df_2007 = df[df['year'] == 2007]

# Identify the countries with populations exceeding 1000 million
df_above_1000 = df_2007[df_2007['population'] > 1000]

# Sort the countries by population in descending order
df_2007_sorted = df_2007.sort_values('population', ascending=False)

# Plotting a bar chart for all countries' populations in 2007
plt.figure(figsize=(12, 8))  # Increase the figure size to accommodate the labels

# Create bars for all countries
bars = plt.bar(df_2007_sorted['country name'], df_2007_sorted['population'], color='lightgreen')

# Annotate China and India with labels
for i, bar in enumerate(bars):
    country_name = df_2007_sorted['country name'].iloc[i]
    if country_name == 'China':
        bar.set_color('green')  # Set color for China
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 50, 'China', rotation=50, 
                 ha='left', va='bottom', color='green', fontsize=12, fontweight='bold')
    elif country_name == 'India':
        bar.set_color('orange')  # Set color for India
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 50, 'India', rotation=50,
                 ha='left', va='bottom', color='orange', fontsize=12, fontweight='bold')

# Add title and labels
plt.title('2007 Global Population Distribution', fontsize=14, color='white')
plt.xlabel('Country', fontsize=12, color='white')
plt.ylabel('Population (Millions)', fontsize=12, color='white')

# Rotate x-axis labels for better readability
plt.xticks(fontsize=9, rotation=90, color='white')

# Set the background color of the entire figure
plt.gcf().set_facecolor('black')
plt.gca().set_facecolor('black')  # Set the axes background to black

# Set y-axis tick color to white
plt.yticks(color='white')

# Add gridlines (optional) to help readability of the bar heights
plt.grid(axis='y', linestyle='--', color='gray', linewidth=0.5)

# Remove the x and y axis lines (spines)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')

# Adjust layout and ensure everything fits
plt.tight_layout()


```
