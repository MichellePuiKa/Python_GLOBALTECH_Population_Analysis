```python
import pandas as pd
from tabulate import tabulate

#Open the CSV file
df = pd.read_csv("/Users/Michelle Wong/Desktop/population.csv")
    
# Reshape the dataset to organize years as columns
df_reshaped = df.pivot(index='country name', columns='year', values='population')

# Filter the data for the year 2000
df_2000 = df[df['year'] == 2000]

# Find countries with a population of 0 in 2000
no_population_2000 = df_2000[df_2000['population'] == 0]

# Display the number of countries with no population data in 2000
print("Number of countries with no population data in 2000: " + str(no_population_2000['country name'].nunique()))

# List the countries with no population data along with their regions
print("Countries with no Population Data in 2000: ")
print(tabulate(no_population_2000[['continent', 'country name']], headers=['Regions', 'Country'], tablefmt='pretty', showindex=False))
```
